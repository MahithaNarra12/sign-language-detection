```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.datasets import mnist
from tensorflow.keras.utils import to_categorical

# Load sample MNIST data (like handwriting digits)
(x_train, y_train), (x_test, y_test) = mnist.load_data()
x_train = x_train / 255.0
x_test = x_test / 255.0
y_train = to_categorical(y_train)
y_test = to_categorical(y_test)

# Create a simple model
model = Sequential([
    Flatten(input_shape=(28, 28)),
    Dense(128, activation='relu'),
    Dense(10, activation='softmax')
])

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
model.fit(x_train, y_train, epochs=2, validation_data=(x_test, y_test))

# Save the model
model.save('model.h5')

# Predict on a sample
import numpy as np
sample = np.expand_dims(x_test[0], axis=0)
pred = model.predict(sample)
print("Predicted class:", np.argmax(pred))
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Example data - replace these with your actual and predicted bike counts
actual_counts = [20, 25, 30, 28, 35, 40, 42, 45, 50, 60]  # Replace with actual values
predicted_counts = [19, 24, 29, 27, 36, 41, 43, 44, 51, 59]  # Replace with predicted values

# Output for comparison
for i in range(len(actual_counts)):
    print(f"Test {i+1}:")
    print(f"  Actual Count: {actual_counts[i]}")
    print(f"  Predicted Count: {predicted_counts[i]}")
    result = "Correct" if actual_counts[i] == predicted_counts[i] else "Incorrect"
    print(f"  Result: {result}")
    print("-" * 40)

# Calculate accuracy
correct_count = sum([1 for a, p in zip(actual_counts, predicted_counts) if a == p])
accuracy = (correct_count / len(actual_counts)) * 100
print(f"Model Accuracy: {accuracy:.2f}%")

# Visualization
plt.figure(figsize=(10, 6))

# Plotting Actual vs Predicted
plt.plot(actual_counts, label="Actual Counts", marker='o', linestyle='-', color='b')
plt.plot(predicted_counts, label="Predicted Counts", marker='x', linestyle='--', color='r')

plt.title("Actual vs Predicted Bike Counts")
plt.xlabel("Time/Instance")
plt.ylabel("Bike Count")
plt.legend()
plt.grid(True)

plt.show()
```
